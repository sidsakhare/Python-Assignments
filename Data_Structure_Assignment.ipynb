1. What are data structures, and why are they important?

 - Data structures are ways of organizing and storing data so it can be accessed and modified efficiently. They are important because they improve program performance, optimize memory usage, and make complex problems easier to solve.

2. Difference between mutable and immutable data types with examples.

 - Mutable data types can be changed after creation.
   Examples: list, dictionary, set

   Immutable data types cannot be changed once created.
   Examples: int, float, string, tuple

3. Main differences between lists and tuples in Python

- Lists are mutable, tuples are immutable

  Lists use [ ], tuples use ( )

  Lists are slower than tuples

  Tuples are safer for fixed data

4. How do dictionaries store data?

- Dictionaries store data as keyâ€“value pairs using a hash table, which allows fast lookup, insertion, and deletion.

5. Why use a set instead of a list?

- Sets automatically remove duplicates and provide faster membership testing compared to lists.

6. What is a string in Python, and how is it different from a list?

- A string is an immutable sequence of characters, while a list is mutable and can store different data types.

7. How do tuples ensure data integrity?

- Tuples are immutable, so their values cannot be changed accidentally, ensuring data remains consistent.

8. What is a hash table, and its relation to dictionaries?

- A hash table is a data structure that maps keys to values using hashing. Python dictionaries are implemented using hash tables.

9. Can lists contain different data types?

- Yes, lists can store multiple data types like integers, strings, floats, and even other lists.

10. Why are strings immutable in Python?

- Strings are immutable for memory efficiency, security, and performance optimization.

11. Advantages of dictionaries over lists

- Faster data retrieval

  Data accessed using meaningful keys

  Better for structured data

12. When is a tuple preferable over a list?

- When data should not be changed, such as   coordinates, database records, or fixed configuration values.

13. How do sets handle duplicate values?

- Sets automatically remove duplicate values and store only unique elements.

14. How does the in keyword differ for lists and dictionaries?

 - In lists: checks for values

   In dictionaries: checks for keys

15. Can you modify elements of a tuple? Why/Why not?

- No, because tuples are immutable, meaning their elements cannot be changed after creation.
