#1. What is NumPy, and why is it widely used in Python?
  NumPy (Numerical Python) is a library used for numerical computing in Python. It provides support for large, multi-dimensional arrays and matrices along with mathematical functions to operate on them. It is widely used because it is fast, memory-efficient, and forms the foundation for many data science and machine learning libraries.

##2. How does broadcasting work in NumPy?
  Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array to match the shape of the larger one without copying data. This enables efficient element-wise operations.

##3.What is a Pandas DataFrame?
  A Pandas DataFrame is a two-dimensional, tabular data structure with labeled rows and columns. It is used for data manipulation and analysis, similar to a spreadsheet or SQL table.

##4. Explain the use of the groupby() method in Pandas.
  The groupby() method is used to split data into groups based on one or more columns, apply aggregation functions (like sum, mean, count), and combine the results.

##5. Why is Seaborn preferred for statistical visualizations?
  Seaborn provides built-in statistical plots, attractive default styles, and integrates well with Pandas DataFrames. It simplifies complex visualizations with fewer lines of code compared to Matplotlib.

##6. What are the differences between NumPy arrays and Python lists?
  NumPy arrays are faster and more memory-efficient.

  Arrays support element-wise operations.

  Lists can store mixed data types, while NumPy arrays store homogeneous data.

  NumPy supports multi-dimensional arrays easily.

##7. What is a heatmap, and when should it be used?
  A heatmap is a graphical representation of data where values are represented by colors. It is commonly used to visualize correlation matrices or intensity of values across categories.

##8. What does the term “vectorized operation” mean in NumPy?
  Vectorized operations perform computations on entire arrays at once without using explicit Python loops, making the operations faster and more efficient.

##9. How does Matplotlib differ from Plotly?
  Matplotlib is a static plotting library mainly used for basic visualizations. Plotly provides interactive and dynamic visualizations with zooming, hovering, and web integration capabilities.

##10. What is the significance of hierarchical indexing in Pandas?
  Hierarchical indexing (MultiIndex) allows multiple levels of indexing on rows or columns. It helps organize complex datasets and perform advanced data analysis.

##11. What is the role of Seaborn’s pairplot() function?
pairplot() creates pairwise scatter plots between numerical variables and shows distributions on the diagonal. It is useful for exploratory data analysis.

##12. What is the purpose of the describe() function in Pandas?
The describe() function generates summary statistics such as count, mean, standard deviation, min, max, and quartiles for numerical columns.

##13. Why is handling missing data important in Pandas?
Missing data can affect analysis accuracy and model performance. Handling missing values ensures clean, reliable datasets for analysis.

##14. What are the benefits of using Plotly for data visualization?
Plotly provides interactive charts, web compatibility, 3D plots, dashboards, and better user engagement compared to static plots.

##15. How does NumPy handle multidimensional arrays?
NumPy uses the ndarray object to represent multi-dimensional arrays. It supports indexing, slicing, reshaping, and mathematical operations across dimensions.

##16. What is the role of Bokeh in data visualization?
Bokeh is a Python library for creating interactive visualizations for web browsers. It is commonly used for dashboards and real-time data applications.

##17. Explain the difference between apply() and map() in Pandas.
map() is used on Series for element-wise transformations.

apply() can be used on both Series and DataFrames and allows applying custom functions across rows or columns.

##20. What is the role of a pivot table in Pandas?

```python
##2.Write a Pandas script to find the mean of a specific column in a DataFrame?
import pandas as pd

A = {'key1':[1,2,3],'key2':[3,4,5],'key3':[4,5,6]}
df =pd.DataFrame(A)
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

study_hours = np.random.randint(1, 10, size=20)
exam_marks = study_hours * 10 + np.random.randint(-10, 10, size=20)

plt.scatter(study_hours, exam_marks)
plt.xlabel("Study Hours")
plt.ylabel("Exam Marks")
plt.title("Study Hours vs Exam Marks")
plt.show()
```

```python
##How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns

data = sns.load_dataset('iris')
```
