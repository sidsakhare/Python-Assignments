18.What is the difference between os.fork() and multiprocessing in Python?

- os.fork()

  Creates a child process by duplicating the current process.

  Available only on Unix/Linux (not Windows).

  Low-level function, harder to manage.

  Child gets an exact copy of parent memory (copy-on-write).

- multiprocessing

  High-level module to create and manage processes easily.

  Works on Windows, Linux, and Mac.

  Provides Process, Pool, Queue, Pipe, etc.

  Safer and more portable for parallel programming.

### Explanation for the previous issue:

The cell with `with open('Loggingfile.log', 'w') as e:` opened the log file in write mode (`'w'`), which truncates the file to zero length or creates it if it doesn't exist. This effectively overwrites any content that the `logging` module might have written or configured to write.

To see the logging output, you need to ensure the logging configuration is set up correctly and the file is not prematurely overwritten or closed by other operations.

Let's try logging again correctly and then display the content of the log file.

```python
import logging


root_logger = logging.getLogger()

for handler in root_logger.handlers[:]:
    root_logger.removeHandler(handler)

    if hasattr(handler, 'close'):
        handler.close()

# To ensure a clean slate for demonstration, we'll clear the log file first (optional, but good for seeing fresh logs)
with open('/content/Loggingfile.log', 'w') as f:
    f.write('')

# Configure logging again. The 'filemode' is 'a' for append, so new logs are added without overwriting.
logging.basicConfig(filename='/content/Loggingfile.log', level=logging.DEBUG,
                    format='%(asctime)s %(levelname)s %(message)s', filemode='a')

logging.debug("This is a debug message from the corrected log.")
logging.info("This is an info message from the corrected log.")
logging.warning("This is a warning message from the corrected log.")
logging.error("This is an error message from the corrected log.")
logging.critical("This is a critical message from the corrected log.")

# It's good practice to call shutdown when you are done with logging to ensure all messages are written.
# However, for a persistent logger throughout a session, you might not call shutdown until the very end.
# For this demonstration, we call it to ensure logs are flushed to disk immediately for verification.
logging.shutdown()

print("Logging messages have been written to Loggingfile.log.")
```

```python
#How would you implement a basic logging setup that logs to a file with rotation after 1MB?

import logging

from logging.handlers import RotatingFileHandler

logger = logging.getLogger('sid')
logger.setLevel(logging.INFO)

FileHandler = RotatingFileHandler('test.log',backupCount = 10, maxBytes=1000000)

FileHandler.namer = lambda name : name.replace('.log', "") + ".log"

logFormatter = logging.Formatter('%(asctime)s-%(levelname)s-%(message)s', datefmt = '%Y-%m-%d %H:%M:%S')

FileHandler.setFormatter(logFormatter)

logger.addHandler(FileHandler)

logger.info("info msg")

try:

  10/"0"
except Exception as e:

  logger.error('some error has occurd', exc_info= True)
```
