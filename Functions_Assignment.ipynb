```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def prime_num():

  for num in range(1, 201):
    if num == 1:
      yield(f"this is the non prime number:{num}")
    if num > 1:
      is_prime = True
      for i in range(2, num):
        if (num % i) == 0:
          is_prime = False
          break

      if is_prime:
        yield(f"{num} is a prime number")
      else:
        yield(f"{num} is not a prime number")
```

```python
A = prime_num()
next(A)

```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fib_series(n):
  a = 0
  b = 1
  for _ in range(n):
    yield a
    a, b = b, a + b
```

```python
Fib_generator = fib_series(10)
next(Fib_generator)
```
